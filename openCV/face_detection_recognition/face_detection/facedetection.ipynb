```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
#loading images
test_image = cv2.imread('download1.jpg')
#converting to grayscale as opencv expects detector takes in input
test_image_gray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)
# displaying grayscale image
plt.imshow(test_image_gray, cmap='gray')
```

```python
# convert 9it to rgb format
def convertToRGB(image):
  return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
uploaded = files.upload()
```

```python
haar_cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(test_image_gray, scaleFactor = 1.2, minNeighbors = 5);

# Let us print the no. of faces found
print('Faces found: ', len(faces_rects))
```

```python
# looping around all the coordinates in the image and draw the rectangle
for (x, y, w, h) in faces_rects:
  cv2.rectangle(test_image, (x, y), (x + w, y + h), (255, 0, 0), 2)
```

```python
# detect the image using the rectangle
plt.imshow(convertToRGB(test_image))
```

```python
# creating a generalised image detector
def detectImage(cascade, test_image, scaleFactor = 1.1):
  image_copy = test_image.copy()
  gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)
  faces_rects = haar_cascade_face.detectMultiScale(gray_image, scaleFactor = scaleFactor, minNeighbors = 5);
  for (x, y, w, h) in faces_rects:
    cv2.rectangle(image_copy, (x, y), (x + w, y + h), (255, 150, 0), 2)

  return image_copy
```

```python
# testing the image on a new image
uploaded = files.upload()
```

```python
test_images_2 = cv2.imread('Untitled-100.jpg')
faces_2 = detectImage(haar_cascade_face, test_images_2)
plt.imshow(convertToRGB(faces_2))
```

```python
test_images_3 = cv2.imread('group_image.jpg')
faces_3 = detectImage(haar_cascade_face, test_images_3)
plt.imshow(convertToRGB(faces_3))
```
